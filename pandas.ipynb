```python
#.apply() method
def calculate_total(row):
    return row['Midterm_Score'] + row['Final_Score'] + row['Assignments_Avg'] + row['Participation_Score'] + row['Total_Score']
df['Total_Score'] = df.apply(calculate_total, axis=1)
df['Total_Score']
# axis1 is column, axis0 is row
```

```python
#using .apply to calculate average score of each student
def calculate_average(row):
    return (row['Midterm_Score'] + row['Final_Score'] + row['Assignments_Avg'] + row['Quizzes_Avg'] + row['Total_Score'] + row['Participation_Score'] + row['Projects_Score'])/7
df['Average_Score'] = df.apply(calculate_average, axis=1)
df['Average_Score']
```

how to categorize students based on average score based on .apply method

```python
def categorize_students(row):
    if row['Average_Score']>= 90:
        return "Excellent"
    elif row['Average_Score']>=75:
        return "Good"
    elif row['Average_Score']>=60:
        return "Average"
    else:
        return "Needs Improvement"
df['Category'] = df.apply(categorize_students, axis=1)
df['Category']
```
