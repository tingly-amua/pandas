```python
#.apply()
def calculate_total(row):
    return row['Midterm_Score'] + row['Final_Score']

df['Total Score'] = df.apply(calculate_total, axis=1)
df['Total Score']
```

```python
df
```

```python
def calculate_average(row):
    return (row['Attendance (%)'] + row['Midterm_Score'] + row['Final_Score'] + row['Assignments_Avg'] + row['Quizzes_Avg'] + row['Participation_Score'] + row['Projects_Score']) / 7

df['Average_Score'] = df.apply(calculate_average, axis=1)
df['Average_Score']
```

```python
df
```

```python
#how to categorize students based on their average score using .apply and .map
def categorize_students(row):
    if row['Average_Score'] >= 90:
        return "Excellent"
    elif row['Average_Score'] >= 75:
        return 'Good'
    elif row['Average_Score'] >= 60:
        return 'Average'
    else:
        return 'Needs Improvement'
    
df['Category'] = df.apply(categorize_students, axis=1)
df['Category']
```

```python
def categorize_score(score):
    if score >= 90:
        return 'Excellent'
    elif score >= 75:
        return 'Good'
    elif score >= 60:
        return 'Average'
    else:
        return 'Needs Improvement'
df['Category'] = df['Average_Score'].map(categorize_score)
```

```python
df
```

```python
def categorize_stress(score):
    if score >= 8:
        return 'Critical Condition'
    elif score >= 6:
        return 'Stressed'
    elif score >= 4:
        return 'Manageable'
    else:
        return 'Live a little!'
df['Stress_Category'] = df['Stress_Level (1-10)'].map(categorize_stress)
```
